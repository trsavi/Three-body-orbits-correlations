```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('finalnaTabela.csv')
```

```python
data.head()
```

```python
#data.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'Index'], inplace=True)
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
def count_letters(df):
    
    df['a'] = df['Slova'].count('a')
    df['b'] = df['Slova'].count('b')
    df['A'] = df['Slova'].count('A')
    df['B'] = df['Slova'].count('B')
    return df

data = data.apply(count_letters, axis=1)
```

```python
data
```

```python
data.drop(data[data['mi']=='-'].index, inplace=True)
```

```python
data['mi'] = data['mi'].astype(float)
```

```python
data
```

```python
data[data['Lf']%2==0]
```

```python
def calc_sym(df):
    df['Diff'] = str(abs(df['A']-df['a']))
    return df
data = data.apply(calc_sym, axis=1)
```

```python
data['Diff'].value_counts()
```

```python
data['Class'] = data['Class'].astype(str)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x = 'v1', y = 'v2', data = data[(data['Stability']=='M') & (data['Class'].isin(['0.5'])) & (data['SubClass'].isin(['A.2']))], hue='Diff')
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x = 'v1', y = 'v2', data = data[(data['Stability']=='M') & (data['Class'].isin(['0.75'])) & (data['SubClass'].isin(['B.2','B.1']))], hue='Diff') #& (data['SubClass'].isin(['A.2']))
# & (data['Diff'].isin(['1']))
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x = 'v1', y = 'v2', data = data[(data['Stability']=='M') & (data['Class'].isin(['4.0'])) & (data['v1']>0.97)& (data['Diff'].isin(['0','1']))], hue='Diff') #& (data['SubClass'].isin(['A.2']))
# & (data['Diff'].isin(['1']))
```

```python
plt.figure(figsize=(15,7))
plot =  sns.scatterplot(x = 'T', y = 'v2', data = data[(data['Stability']=='M') & (data['Class'].isin(['0.75']))  & (data['SubClass'].isin(['B.1','B.2']))], hue='Lf',  palette='hls')
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.xlabel('T', fontsize=14)
plt.ylabel('v2', fontsize=14)
plt.setp(plot.get_legend().get_texts(), fontsize='12') # for legend text
plt.setp(plot.get_legend().get_title(), fontsize='13') # for legend title
plt.show()
```
